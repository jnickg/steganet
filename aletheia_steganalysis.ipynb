## Comparison of encoded images

```python
import matplotlib.pyplot as plt
ifile = 'steganet/assets/test_images/input.png'
bfile = 'steganet/assets/test_images/Basic_output.png'
dfile = 'steganet/assets/test_images/Dense_output.png'

fig, ax = plt.subplots(1, 3, figsize=(16,4))

fig.suptitle('Comparison of Encoded Images')

img = plt.imread(ifile)
ax[0].imshow(img)
ax[0].set_title('No Encoding')

img = plt.imread(bfile)
ax[1].imshow(img)
ax[1].set_title('Basic Encoding')

img = plt.imread(dfile)
ax[2].imshow(img)
ax[2].set_title('Dense Encoding')

plt.show()
```

As you can see in the figure above, the encoding of the source image (left) using a basic encoding (middle), changes the coloring of the image in a rather perceptible way. The dense encoding (right) also makes noticible changes to the appearance of the source image; however, the changes are less dramatic and thus, less noticible than the basic encoding.

## Image Histograms
Below you can see the RGB color distributions and intensities for each image.

```python
import imageio
import aletheia.aletheialib.utils

def plot_img_hist(fn, title):
  I = imageio.imread(fn)
  data = []
  if len(I.shape) == 1:
      data.append(I.flatten())
  else:
      for i in range(I.shape[2]):
          data.append(I[:,:,i].flatten())

  plt.hist(data, range(0, 255), color=["r", "g", "b"])
  plt.suptitle(title)
  plt.show()
```

```python
plot_img_hist(fn='steganet/assets/test_images/input.png', title='Source Image')
```

```python
plot_img_hist(fn='steganet/assets/test_images/Basic_output.png', title='Basic Encoding')
```

```python
plot_img_hist(fn='steganet/assets/test_images/Dense_output.png', title='Dense Encoding')
```

In comparison to the source image, we observe that the Basic Encoding mutes the color channels, and the color values shift more towards the middle range. For example, in the source image, the blue channel is most concentrated in the 200 to approximately 230 range, whereas in the basic encoding histogram, it appears to congregate aaround the 150 to 170 range.

Similarly, for the dense encoding, the color values tend to shift towards lower intensity; however, the shift is less dramatic in comparison to the basic encoding. In addition to this, the channels appear to be flattened slightly, which likely is the reason we observe the slight lowering of contrast between the source image and the densely encoded image.
