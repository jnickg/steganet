## Testings multiple models at once, using `auto` command on our images.

The first image, is a baseline image with no steganographic encoding. As expected, non of the models / algorithms detected the presence of any steganographic encodings.

```python
!python aletheia/aletheia.py auto steganet/assets/test_images/input.png
```

Next, the input image `Basic_output.png` contains a basic encoding.  
As you can see in the results below, SteganoGAN performed the worst, with only a 10% probability of an encoding being present. LSBM performed the best with a score of 90%, and HILL performed the second best, with a probability of 70%.

```python
!python aletheia/aletheia.py auto steganet/assets/test_images/Basic_output.png
```

Lastly, the input image `Dense_output.png` contains a dense encoding.  
As you can see in the results below, again, SteganoGAN performed the poorest, this time with a 20% probability of an encoding being present. LSBM again performed the best, giving a 100% probability of there being a steganographic encoding, and HILL is in second place, with a probability score of 90%.

```python
!python aletheia/aletheia.py auto steganet/assets/test_images/Dense_output.png
```
